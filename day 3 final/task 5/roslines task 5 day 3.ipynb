```python
import matplotlib.pyplot as plt

# Sample Data (income vs life expectancy for several countries across different years)
countries = ['Russia', 'China', 'India', 'Ireland', 'Australia']
years = [2000, 2010, 2020]  # Simulating data for three years
income_data = {
    2000: [35000, 3000, 800, 40000, 10000],  # Average income per country in 2000
    2010: [40000, 7000, 1500, 45000, 15000],  # Average income per country in 2010
    2020: [45000, 10000, 2000, 50000, 20000]  # Average income per country in 2020
}
life_expectancy_data = {
    2000: [78, 70, 64, 81, 73],  # Life expectancy for each country in 2000
    2010: [79, 74, 68, 82, 74],  # Life expectancy for each country in 2010
    2020: [80, 76, 70, 83, 75]   # Life expectancy for each country in 2020
}

# Choose the year for the static chart (e.g., 2020)
year = 2010

# Create a bubble chart
plt.figure(figsize=(10, 6))
bubble_size = [i * 10 for i in income_data[year]]  # Bubble size based on income

# Plot the data
plt.scatter(income_data[year], life_expectancy_data[year], s=bubble_size, alpha=0.6)

# Add title and labels
plt.title(f'Life Expectancy vs Income in {year}')
plt.xlabel('Income per Person (USD)')
plt.ylabel('Life Expectancy (Years)')

# Annotate country names on the plot
for i, country in enumerate(countries):
    plt.annotate(country, (income_data[year][i], life_expectancy_data[year][i]), fontsize=12, ha='right')

# Display the chart
plt.show()
```
