```python
import matplotlib.pyplot as plt

# Data for the chart (number of cell phones per person in different countries)
countries = ['USA', 'China', 'India', 'Germany', 'Brazil']
cell_phones_per_person = [1.1, 0.9, 0.7, 1.3, 0.8]  # Example data

# Create a line chart
plt.plot(countries, cell_phones_per_person, marker='o', color='green', linestyle='-', linewidth=2, markersize=8)

# Add title and labels
plt.title('Number of Cell Phones Per Person by Country')
plt.xlabel('Country')
plt.ylabel('Cell Phones Per Person')

# Display the chart
plt.show()
```
