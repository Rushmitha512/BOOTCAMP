```python
import matplotlib.pyplot as plt

# Data for the chart
countries = ['Finland', 'Denmark', 'Switzerland', 'Iceland', 'Netherlands']
happiness_scores = [7.8, 7.6, 7.5, 7.5, 7.4]

# Create a line chart
plt.plot(countries, happiness_scores, marker='o', color='blue', linestyle='-', linewidth=2, markersize=8)

# Add title and labels
plt.title('Happiness Scores by Country')
plt.xlabel('Country')
plt.ylabel('Happiness Score')

# Display the chart
plt.show()

```
